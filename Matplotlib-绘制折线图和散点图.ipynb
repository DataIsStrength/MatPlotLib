Matplotlib-绘制折线图和散点图

```python
'''
Matplotlib三层结构：
1.容器层：
（1）画板层；
（2）画布层：plt.figure()；
（3）绘图区/坐标系:plt.subplots()；
2.辅助显示层：标题、网格、图例等；
3.图像层：图形的风格，包括折线图、散点图、柱状图等。
'''
```

```python
#导入Matplotlib包
import matplotlib.pyplot as plt
#在一些运行环境下需要输入此行
%matplotlib inline

#展现上海一周的天气,比如从星期一到星期日的天气温度如下
#创建画布，dpi参数为图像的清晰度
plt.figure(figsize=(20,8),dpi=80)

#绘制图像
plt.plot([1,2,3,4,5,6,7],[17,17,18,15,11,11,13])

#保存图像,注意要在plt.show()之前保存，也可以直接右键保存
#plt.savefig('保存路径')

#显示图像
plt.show()
```

```python
#画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度

#导入random包
import random

#正常显示
#正常显示中文
plt.rcParams['font.sans-serif']=['SimHei']
#正常显示负号
plt.rcParams['axes.unicode_minus']=False

#准备数据
x=range(60)
y_shanghai=[random.uniform(15,18) for i in x]

#创建画布
plt.figure(figsize=(20, 8),dpi=80)

#绘制图像
plt.plot(x,y_shanghai)

#辅助显示层
#设置x的刻度说明
x_label=['11点{}分'.format(i) for i in x]
#设置x和x_label的步长为5
plt.xticks(x[::5],x_label[::5])
#y轴刻度为[0,40)步长为5
plt.yticks(range(0,40,5))
#添加网格显示，alpha参数为网格透明度
plt.grid(linestyle='--',alpha=0.5)
#添加轴标签和标题
plt.xlabel('时间变化')
plt.ylabel('温度变化')
plt.title('某城市11点到12点每分钟的温度变化状况')

#显示图像
plt.show()
```

```python
#再添加一个城市的温度变化

#准备数据
x=range(60)
y_shanghai=[random.uniform(15,18) for i in x]
y_beijing=[random.uniform(1,3) for i in x]

#创建画布
plt.figure(figsize=(20,8),dpi=80)

#绘制图像
plt.plot(x,y_shanghai,color='m',linestyle='-.',label='上海')
plt.plot(x,y_beijing,color='c',label='北京')
'''
图形风格参数：
（1）颜色：r红色、g绿色、b蓝色、w白色、c青色、m洋红、y黄色、k黑色；
（2）线条：-实线、--虚线、-.点划线、:点虚线。
'''

#显示图例，位置参数loc默认值为0，表示最佳位置
plt.legend(loc=0)

#辅助显示层
#设置x的刻度说明
x_label=['11点{}分'.format(i) for i in x]
#设置x和x_label的步长为5
plt.xticks(x[::5],x_label[::5])
#y轴刻度为[0,40)步长为5
plt.yticks(range(0,40,5))
#添加网格显示
plt.grid(linestyle='--',alpha=0.5)
#添加轴标签和标题
plt.xlabel('时间变化')
plt.ylabel('温度变化')
plt.title('上海、北京11点到12点每分钟的温度变化状况')

#显示图像
plt.show()
```

```python
#在同一个画布绘制多个图像

#准备数据
x=range(60)
y_shanghai=[random.uniform(15,18) for i in x]
y_beijing=[random.uniform(1,3) for i in x]

#创建画布
plt.figure(figsize=(20,8),dpi=80)

#绘制左侧图
#设置图片位置为1行2列的第1个位置
plt.subplot(121)
plt.plot(x,y_shanghai,color='m',linestyle='-.',label='上海')
plt.legend()
x_label=['11点{}分'.format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(0,40,5))
plt.grid(linestyle="--", alpha=0.5)
plt.xlabel('时间变化')
plt.ylabel('温度变化')
plt.title('上海11点到12点每分钟的温度变化状况')

#绘制右侧图
#设置图片位置为1行2列的第2个位置
plt.subplot(122)
plt.plot(x,y_beijing,color='c',label='北京')
plt.legend()
x_label=['11点{}分'.format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(0,40,5))
plt.grid(linestyle='--',alpha=0.5)
plt.xlabel('时间变化')
plt.ylabel('温度变化')
plt.title('北京11点到12点每分钟的温度变化状况')

#显示图像
plt.show()
```

```python
#绘制函数图像

#导入NumPy包
import numpy as np

#准备数据
#生成等差数组
x=np.linspace(-1,1,1000)
y=x**2

#创建画布
plt.figure(figsize=(20,8),dpi=80)

#绘制图像
plt.plot(x,y)

#添加网格显示
plt.grid(linestyle='--',alpha=0.5)

#显示图像
plt.show()
```
